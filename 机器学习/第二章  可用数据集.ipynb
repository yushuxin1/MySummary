```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

def datasets_demo():
    """
    sklearn数据集使用
    return
    """
    #   获取数据集
    iris = load_iris()
    print("鸢尾花数据集：\n", iris)
    print("查看数据集描述：\n", iris["DESCR"])
    print("查看特征值名字：\n", iris.feature_names)
    print("查看特征值：\n", iris.data.shape)
    
    #  数据集划分
    x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=22)
    print("训练集的特征值：\n", x_train, x_train.shape)
    
    return None

if __name__ == "__main__":
    #   sklearn数据集使用
    datasets_demo()
```

### 2.5特征提取
sklearn.feature.extraction  
#### 1 字典特征提取 - 类别 =》one-hot编码  
&nbsp;&nbsp; sklearn.feature.extraction.DictVectorizer(sparse=True)    vector 向量   矩阵(matrix，二维数组)  
&nbsp;&nbsp; &nbsp;&nbsp;  返回sparse矩阵
&nbsp;&nbsp; sparse稀疏矩阵  将非零值按位置表示出来      节省内存-提高加载效率
&nbsp;&nbsp; 应用场景：  
&nbsp;&nbsp; &nbsp;&nbsp;1） 数据集当中类别特征比较多  
&nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; 将数据特征 -》字典特征  
&nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp; DictVctorizer转换  
&nbsp;&nbsp; &nbsp;&nbsp;2）本身拿到的数据就是字典类型  
#### 2 文本特征提取
&nbsp;&nbsp;  单词 作为 特征
&nbsp;&nbsp;  特征：特征词
&nbsp;&nbsp;方法一： CountVectorizer
&nbsp;&nbsp;&nbsp;&nbsp;统计每个样本特征值出现的次数
&nbsp;&nbsp;&nbsp;&nbsp;stop_words:停用的  停用词表

```python
from sklearn.feature_extraction import DictVectorizer
def dict_demo():
    """
    字典特征提取
    """
    data = [
        {'city': '北京', 'temperature': 100},
        {'city': '上海', 'temperature': 60},
        {'city': '深圳', 'temperature': 30}
    ]
    #  1、实例化一个转换器
    trander = DictVectorizer(sparse=True)   #或者直接为sparse=False data_new不用加.toarray()
    #  2、调用fit_transform()
    data_new = trander.fit_transform(data)
    print('data_new:\n', data_new.toarray())
    print('特征名字：\n', trander.get_feature_names())
    return None
dict_demo()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
def count_demo():
    """
    文本特征抽取: CountVectorizer
    """
    data = ['life is short, i like like python', 'life is too long, i dislike %%python']
    #  1、实例化一个转换器类
    transfer = CountVectorizer()
    #  2、调用fit_transform()
    data_new = transfer.fit_transform(data)
    print('data_newata:\n', data_new.toarray())
    print('特征名字：\n', transfer.get_feature_names())
    return None
count_demo()
```
