```python
import face_recognition
import cv2
```

```python
# 打开视频， 参数是0即VideoCapture(0)打开笔记本的内置摄像头，参数是视频路径则打开该视频
video_capture = cv2.VideoCapture("short_hamilton_clip.mp4")
```

```python
# frames不存图片的每一帧
frames = []
frame_count = 0
```

```python
# 如果视频是在打开状态下
while video_capture.isOpened():
    #  抓住一帧视频，ret是一个布尔值，如果读取的图像是正确的就返回True，读取到文件结尾是False
    #  frame是每一帧的图像是一个三维矩阵
    ret, frame = video_capture.read()
    
    #  视频文件结束时退出
    if not ret:
        break
        
    #  将图像从BGR颜色转化为RGB颜色
    frame = frame[:, :, ::-1]
    
    #  将视频中的每一帧保存到列表中
    frame_count += 1
    frames.append(frame)
    
    # 每128帧（默认批量），批处理帧以查找到面部
#     使用cnn面部检测器返回图像中人脸边界框的二维数组如果您使用的是GPU，这可以为您提供更快的结果，因为GPU可以一次处理批量图像
    if len(frames) == 128:
        batch_of_face_locations = face_recognition.batch_face_locations(frames, number_of_times_to_upsample=0)
        
        #  现在让我们列出我们在所有128帧中找到的所有面孔
        for frame_number_in_batch, face_locations in enumerate(batch_of_face_locations):
            number_of_faces_in_frame = len(face_locations)
            
            #  如果使用frame_number_in_batch到达128帧时又会重新开始计时
            frame_number = frame_count - 128 + frame_number_in_batch
            print("I found {} face(s) in frame #{}".format(number_of_faces_in_frame, frame_number))
            
            for face_location in face_locations:
                #  打印此框中每个脸的位置
                top, right, bottom, left = face_location
                print("一个面部位于像素位置： top:{}, right:{}, bottom:{}, left:{}".format(top, right, bottom, left))
            
        #  清除帧阵列以启动下一批
        frames = []
```
